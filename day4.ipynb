```python
with open ('input/input4.txt') as f:
    data = [x.rstrip() for x in f.readlines()]
```

```python
numbers = [int(n) for n in data[0].split(',')]
```

```python
board = -1
boards = []
for i in range(1, len(data)):
    if data[i] == '':
        board += 1
        boards.append([])
    else:
        boards[board].append([int(n) for n in data[i].split(' ') if n != ''])
```

```python
def play_game(number, boards):
    for i in range(len(boards)):
        for j,r in enumerate(boards[i]):
            boards[i][j] = [n if n != number else -1 for n in r]
```

```python
def check_game():
    for i, bo in enumerate(boards):
        if -5 in [sum(row) for row in bo]:
            print(boards.pop(i))
        elif -5 in [sum([row[j] for row in bo]) for j in range(5)]:
            print(boards.pop(i))
    return -1
```

```python
for num in numbers: 
    play_game(num, boards)
    check_game()
    print(num)
```
